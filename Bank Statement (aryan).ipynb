```python
test2['Amount (Rs.)']=test2['Amount (Rs.)'].astype('str')
```

```python
idx =[]
for i in range(test2.shape[0]):
    if "Amount (Rs.)" in test2.iloc[i,3]:
        idx.append(i)
```

```python
test2 = test2.drop(idx,axis=0)
test2
```

```python
test2.drop([154,155],axis=0,inplace=True)
```

```python
test2.shape
```

```python
cr_list,db_list =[],[]
```

```python
for i in range(test2.shape[0]):
    if "Dr" in test2.iloc[i,3]:
        db_list.append(test2.iloc[i,3])
        cr_list.append("NaN")
    else:
        db_list.append("NaN")
        cr_list.append(test2.iloc[i,3])

```

```python
print(len(db_list))
print(len(cr_list))
print(db_list)
print(cr_list)
```

```python
db_list2,cr_list2=[],[]

for i in db_list:
    if "Dr" in i:
        db_list2.append(i[:-5])
    else:
        db_list2.append(i)

for i in cr_list:
    if "Cr" in i:
        cr_list2.append(i[:-5])
    else:
        cr_list2.append(i)
```

```python
db_list2
```

```python
cr_list2
```

```python
x = test2['Balance (Rs.)']
```

```python
df2 = pd.DataFrame({"Tran ID":test2['Tran Id'],"Debit amount":db_list2,"Credit amount":cr_list2,"Balance":x })
```
